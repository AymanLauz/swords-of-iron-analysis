````python
# Clean the Timeline
df_alarms['time'] = pd.to_datetime(df_alarms['time'],
                                   errors='coerce')

# Drop any rows that had bad dates
df_alarms = df_alarms.dropna(subset=['time'])

print("Oldest alarm in data:", df_alarms['time'].min())
print("Newest alarm in data:", df_alarms['time'].max())
````

````python
# Picking the segment of dates from October the 7th 2023 up till today
start_date = "2023-10-07"
end_date = "2025-10-16"

df_alarms_war_period = df_alarms[
    (df_alarms['time'] >= start_date) &
    (df_alarms['time'] <= end_date)
].copy()

print(f"Total alarms in original df_alarms: {df_alarms.shape[0]}")
print(f"Total alarms in filtered df_alarms_war_period: {df_alarms_war_period.shape[0]}")

print("\nFirst 5 alarms:")
print(df_alarms_war_period.head())

print("\nLast 5 alarms:")
print(df_alarms_war_period.tail())
````

# Analysis 1: Most & Least Affected Cities

````python
print("--- [Module 1C] Analysis 1: City Alarm Counts ---")

# 1. Get the counts of alarms for every unique city
# This counts occurrences of each city name in the 'cities' column
city_alarm_counts = df_alarms_geocoded['cities'].value_counts()

# 2. Print the top 10 most frequent
print("\n--- Top 10 Most Alarms ---")
print(city_alarm_counts.head(10))

# 3. Print the 10 least frequent (for cities that had at least one)
print("\n--- 10 Cities with Fewest Alarms ---")
print(city_alarm_counts.tail(10))

# 4. Show how many cities had only one alarm
cities_with_one_alarm = city_alarm_counts[city_alarm_counts == 1].shape[0]
print(f"\nNumber of cities with exactly 1 alarm: {cities_with_one_alarm}")
````

# Analysis 2: "Zoom In" on a Specific Date

````python
import folium
from folium.plugins import HeatMap
import os
import pandas as pd

print("\n--- [Module 1C] Analysis 2: Specific Date 'Zoom-In' ---")

# Ensure the 'time' column is datetime type
df_alarms_geocoded['time'] = pd.to_datetime(df_alarms_geocoded['time'])

# --- 1. Define the date and filter ---
specific_date_str = '2025-06-13' # You can change this date
specific_date = pd.to_datetime(specific_date_str).date()

# Filter the DataFrame for this day
df_specific_day = df_alarms_geocoded[df_alarms_geocoded['time'].dt.date == specific_date]

# --- 2. Print a summary ---
print(f"\n--- Summary for {specific_date_str} ---")
print(f"Total alarms: {len(df_specific_day)}")
print(f"Unique cities affected: {df_specific_day['cities'].nunique()}")
print("\nThreat origins on this day:")
print(df_specific_day['origin'].value_counts())
print("\nThreat types on this day:")
print(df_specific_day['description'].value_counts())

# --- 3. Create a new map just for this day ---
if len(df_specific_day) > 0:
    print("\nCreating a specific heatmap for this day...")
    map_center = [31.0, 34.8]
    m_day = folium.Map(location=map_center, zoom_start=8)

    # Prepare data (filter out any remaining NaNs)
    heat_data_day = df_specific_day.dropna(subset=['coordinates'])['coordinates'].tolist()

    HeatMap(heat_data_day).add_to(m_day)

    # Save the map
    maps_dir = path_to_folder + "/maps"
    os.makedirs(maps_dir, exist_ok=True) # Ensure folder exists
    map_save_path_day = maps_dir + f"/alarms_on_{specific_date_str}.html"
    m_day.save(map_save_path_day)

    print(f"Map saved to: {map_save_path_day}")
else:
    print("No alarms found for this date.")
````

most dangerous time of day?

````python
import matplotlib.pyplot as plt

print("\n--- [Module 1C] Analysis 3A: Alarms by Hour of Day ---")

# Extract the hour from the 'time' column
alarms_by_hour = df_alarms_geocoded['time'].dt.hour.value_counts().sort_index()

print(alarms_by_hour)

# Plot a bar chart
plt.figure(figsize=(12, 6))
alarms_by_hour.plot(kind='bar')
plt.title('Total Alarms by Hour of Day (0-23)')
plt.xlabel('Hour of Day')
plt.ylabel('Total Alarms')
plt.grid(axis='y', linestyle='--')
plt.show()
````

What was the timeline of the war?

````python
print("\n--- [Module 1C] Analysis 3B: Alarms Per Day (Timeline) ---")

# Group by date and count alarms
# We must use .set_index('time') for .resample() to work
alarms_per_day = df_alarms_geocoded.set_index('time').resample('D').size()

print(alarms_per_day.head())

# Plot a line chart
plt.figure(figsize=(15, 6))
alarms_per_day.plot()
plt.title('Total Alarms Per Day (2023-2025)')
plt.xlabel('Date')
plt.ylabel('Total Alarms')
plt.grid(True)
plt.show()
````
